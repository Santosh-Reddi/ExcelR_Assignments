```python
# for small samples (<50) we use t-statistics
# n = 25, degree of freedom = 25-1 = 24
# for 95% confidence interval, alpha = 1% = 0.01 and alpha/2 = 0.005
from scipy import stats

ci = 95
n = 25
tscore = stats.t.ppf(1- ((100-ci)/2/100), n-1) # 95% CI, t8,0.005
print(tscore)
```
