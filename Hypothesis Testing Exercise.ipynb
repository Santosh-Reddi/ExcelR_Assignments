# Cutlet

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df=pd.read_csv("Cutlets.csv")
```

```python
df.head()
```

```python
#Null Hypothessis : There is no significant difference in size of cutlet prodcuced by Unit-A and Unit-B,Two means are equal
#ALternate Hypothesis : There's a significant difference between Unit-A and Unit-B cutlet size,Two means are not equal
#This is two sample two tail test
```

```python
df.shape
```

```python
import seaborn as sns
```

```python
sns.distplot(df.iloc[:,0])
```

```python
sns.distplot(df["Unit A"])
```

```python
sns.kdeplot(df["Unit A"])
sns.kdeplot(df["Unit B"])
```

```python
print('Average cutlet size from Unit-A is',round(np.average(df["Unit A"]),2))
print('Average cutlet size from Unit-B is',round(np.average(df["Unit B"]),2))
```

```python
stats.ttest_ind(df.iloc[:,0],df.iloc[:,1])
```

```python
#Since p>(Alpha:0.05) , we accept Null Hypothesis(Ho),
#i.e, There is no significant difference in size of cutlet prodcuced by Unit-A and Unit-B
```

# LabTAT

```python
df=pd.read_csv("LabTAT.csv")
```

```python
df.head()
```

```python
#Comparison of variation between more than two variable,hence we use ANOVA
#Null Hypothesis(Ho) : There is no variation/difference in average TAT
#Alternate Hypothesis (Ha): There's some difference in TAT
```

```python
stats.f_oneway(df.iloc[:,0],df.iloc[:,1],df.iloc[:,2],df.iloc[:,3])
```

```python
#Since p< (Alpha: 0.05) and F value is high, we reject Null Hypothesis(Ho)
```

# Customer Order Form

```python
df=pd.read_csv("Costomer+OrderForm.csv")
```

```python
df.head()
```

```python
#Null Hypothesis(H0): Percentage of defects doesn't varies between the centers
#Alternate Hypothesis(Ha):Percentage of defects varies between the centers
#Since there are more than two variables to compare, we use ANOVA
```

```python
df.replace({'Error Free': 0, 'Defective':1}, inplace=True)
```

```python
(a,b)=stats.f_oneway(df.iloc[:,0],df.iloc[:,1],df.iloc[:,2],df.iloc[:,3])
```

```python
if b<0.05:
    print("We reject Null Hypothesis(H0)")
else:
    print("We accept Null Hypothesis(H0)")
```
